

$$
S(x) = \frac{1}{1+e^{-x}}
$$


$$
\frac{d}{dx}S(x) = \frac{d}{dx} \left [ \frac{1}{1+e^{-x}} \right ] =  \frac{e^{-x}}{\left ( 1+e^{-x} \right )^2} = e^{-x}S^2(x)
$$

$$
\frac{dS}{dx} = \frac{e^{-x}}{1+e^{-x}}\frac{1}{1+e^{-x}} = \frac{e^{-x}\overbrace{+1-1}^{+0}}{1+e^{-x}}\frac{1}{1+e^{-x}} = \frac{(1+e^{-x})-1}{1+e^{-x}}\frac{1}{1+e^{-x}} 
$$

$$
 \frac{(1+e^{-x})-1}{1+e^{-x}} =  \frac{1+e^{-x}}{1+e^{-x}} -  \frac{1}{1+e^{-x}} =  1 -  \frac{1}{1+e^{-x}}
$$

$$
\frac{dS}{dx} = \left ( 1 -  \overbrace{\frac{1}{1+e^{-x}}}^{S(x)} \right ) \overbrace{\frac{1}{1+e^{-x}}}^{S(x)} = (1-S(x))S(x)
$$

$$
\frac{dS}{dx} = S'\left(S\right) = S(1-S) 
$$
[Derivative of the Sigmoid function, Arunava](https://towardsdatascience.com/derivative-of-the-sigmoid-function-536880cf918e)

Consider $N$ layers (input, output and hidden), with $M_l$ neurons in $l$-th layer, $x_{ml}$ the input and $b_{ml}$ the bias of the $m$-th neuron at layer $l$, with input weight $\omega_{jml}$ (forming the entire set of weights $\Omega$) multiplying the output from neuron $j$ to neuron $m$ at layer $l$ and with output $y_{ml}$.

The entire network may be written as:
$$
Y_N = \Gamma(X_1)
$$


$$
\begin{bmatrix} y_{1N} \\ \vdots \\ y_{M_NN} \end{bmatrix} = \Gamma \left( \begin{bmatrix} x_{11} \\ \vdots \\ x_{M_11} \end{bmatrix} \right) 
$$

The last layer is the output layer. So, for a given sample $\bar{X}_c$ with $M_1$ features we have a class $c | \bar{X}_c \in C_c$ and $c = 1,...,M_N$. So, we must have the number of input neurons as the number of sample features and the same number of outputs as the number of the classes. So, if $\bar{X}_c \in C_c$ we want predict the class $c$ with $y_{cN} = 1$ and $y_{mN} = 0 ~~\forall ~~m \neq c$. In other words, we want activate only the output neuron of the respective class.

With that, our optimization problem is:
$$
\Omega^{*} = \arg\min_{\Omega} \sum\limits^{K}_{k=1} \left|\bar{Y}^{(k)}_{c} - Y^{(k)}_{N}\right|
$$
for a set of $K$ samples with true vector class $\bar{Y}_{c}$.

$$
E^{(k)} \triangleq \left \|\bar{Y}^{(k)}_{c} - Y^{(k)}_{N} \right \|^2_2 = \left\|\bar{Y}^{(k)}_{c} - \Gamma(\bar{X}^{(k)}_c)\right\|^2_2
$$

$$
E = \begin{bmatrix} e_{1} \\ \vdots \\ e_{M_N} \end{bmatrix}
$$

Using NewtonRaphson method, we may achieve the $\Omega^*$ iterating over $\omega_{jml}$:

$$
\omega_{jml} \leftarrow \omega_{jml} - \eta \frac{\partial e_q}{\partial \omega_{jml} }
$$

$$
y = \phi(t = w^Tx) ~~\text{and}~~ e = y-y_{True}
$$

$$
\frac{\partial e}{\partial w} = \frac{\partial e}{\partial y}\frac{\partial y}{\partial w} = \frac{\partial e}{\partial y}\frac{\partial y}{\partial t}\frac{\partial t}{\partial w} 
$$



$$
\frac{\partial e}{\partial y} = 1 \\
\frac{\partial y}{\partial x} = \frac{\partial \phi(t)}{\partial t} \\
\phi(t) = S(t) \\
\frac{\partial S(t)}{\partial t} = S(t)(1-S(t)) ~~\therefore~~ \frac{\partial y}{\partial x} = S(1-S) \Big |_{t=w^Tx} \\
u \triangleq \frac{w}{\|w\|} \\
\frac{\partial t}{\partial w} = \frac{\partial}{\partial w}\left[ w^Tx\right] = u^Tx = \sum\limits_{i} x_i
$$


$$
\frac{\partial e}{\partial w} = \overbrace{\left( 1 \right)}^{\frac{\partial e}{\partial y}} \underbrace{\left( S(1-S)\right)}_{\frac{\partial y}{\partial t}} \overbrace{\left( \sum\limits_{i} x_i \right)}^{\frac{\partial t}{\partial w}}
$$

$$
\frac{\partial e}{\partial w} = S(1-S)\sum\limits_{i} x_i
$$

$$
w \leftarrow  w - S(1-S)\sum\limits_{i} x_i
$$

In case of a sequence of layers: $x_i = o_i$, the output of the neurons from the previous layer.

Conclusion: how close the prediction is of to the true value, smaller is the error. The error function is convex and have one only root, at $E(\cdot) = 0$:
$$
E(\Omega) = 0 ~~\therefore~~ \frac{\partial E}{\partial\Omega} = 0
$$

$$
e = 0 ~~\therefore~~ \frac{\partial e}{\partial\omega} = 0 ~~\therefore~~ w \leftarrow  w - 0
$$

With that, the weights is not more updated for new samples, the network has been learned.

```python
%%writefile util.py

from typing import List
from math import exp

def dot_prod(xs: List[float], ys: List[float]) -> float:
    return sum(x*y for x,y in zip(xs,ys))

def sigmoid(x: float) -> float:
    return 1.0/(1.0 +exp(-x))

def derivative_sigmoid(x: float) -> float:
    sig = sigmoid(x)
    return sig*(1 - sig)
```

```python
%%writefile neuron.py

from typing import List, Callable
from util import dot_prod

class Neuron:
    def __init__(self, weights: List[float], learning_rate: float,
                 activation_function: Callable[[float], float],
                 derivative_activation_function: Callable[[float], float]):
        self.weights = weights
        self.activation_function = activation_function
        self.derivative_activation_function = derivative_activation_function
        self.learning_rate = learning_rate
        self.output_cache = 0.0
        self.delta = 0.0
        
    def output(self, inputs: List[float]) -> float:
        self.output_cache = dot_prod(inputs, self.weights)
        return self.activation_function(self.output_cache)
    
    
```

```python
%%writefile layer.py

from __future__ import annotations
from typing import Lista, Callable, Optional
from random import random
from neuron import Neuron
from util import dot_prod

class Layer:
    def __init__(self, previous_layer: Optional[Layer], num_neurons: int,
                learning_rate: float, activation_function: Callable[[float], float],
                derivative_activation_function: Callable[[float], float]):
        self.previous_layer = previous_layer
        self.neurons: List[Neuron] = []
        for i in range(num_neurons):
            if previous_layer is None:
                random_weights = []
            else:
                random_weights = [random() for _ in range(len(previous_layer.neurons))]
            neuron = Neuron(random_weights, learning_rate, activation_function,
                           derivative_activation_function)
            self.neurons.append(neuron)
        self.output_cache = [0.0 for _ in range(num_neurons)]

    def outputs(self, inputs: List[float]) -> List[float]:
        if self.previous_layer is None:
            self.output_cache = inputs
        else:
            self.output_cache = [n.output(inputs) for n in self.neurons]
        return self.output_cache
    
    def calculate_deltas_for_output_layer(self, expected: List[float]):
        for n in range(len(self.neurons)):
            tmp = self.neurons[n].output_cache * (expected[n] - self.output_cache[n])
            self.neurons[n].delta = self.neurons[n].derivative_activation_function(tmp)
            
    def calulate_deltas_for_hidden_layer(self, next_layer):
        for index, neuron in enumerate(self.neurons):
            next_weights = [n.weights[index] for n in next_layer.neurons]
            next_deltas = [n.delta for n in next_layer.neurons]
            sum_weights_and_deltas = dot_prod(next_weights, next_deltas)
            neuron.delta = neuron.derivative_activation_function(neuron.output_cache)*sum_weights_and_deltas
            
```
